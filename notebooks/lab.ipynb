```python
(
    summary,
    history,
    divisions, 
    municipalities
) = uy.fetch_data(country)
```

```python
from IPython.display import display
from rapidfuzz.fuzz import token_sort_ratio as tsr
```

```python
display(divisions)
display(municipalities)
```

```python
dfd = divisions.copy()
dfm = municipalities.copy()
```

```python
dfd["_muni_dept"] = dfm["department"].copy().unique()
dfd 
```

```python
def tokenize(row, pairs):
    for p in pairs:
        token = tsr(row["department"], p)
        if token > 85:
            return p
    return pd.NA
```

```python
dfd["_dept"] = dfd.apply(lambda row: tokenize(row, dfd["_muni_dept"].to_list()), axis=1)
dfd
```

```python
dfm = dfm.rename(columns={"department":"_dept"})
dfm
```

```python
df = (
    dfm
    .merge(dfd, how="left", on="_dept")
    .rename(
        columns={
            "population_x": "municipality_ppl",
            "surface": "municipality_area_km2",
            "creation": "municipality_creation_date",
            "capital": "province_capital",
            "department": "province",
            "area_km2": "province_area_km2",
            "population_y": "province_ppl"
        }
    )
    .assign(
        country_code="UY",
    )
    .drop(
        columns=["_dept", "mayor", "area_mi2", "_muni_dept"],
        axis=1
    )
)
df["province_code"] = (
    df["province"]
    .str.replace(r"A|a|E|e|I|i|O|o|U|u|\s+", "", regex=True)
    .str.upper()[:4]
    .str.replace("Í", "I", regex=False)
)
df
```

```python
dft = tuy.transform(divisions, municipalities)
dft
```

```python
import numpy as np
```

```python
dft["municipality_area_km2"] = (
    dft["municipality_area_km2"]
    .replace(r"\s+km.*", " ", regex=True)
    .replace(r"\(sin\sdatos\)", pd.NA, regex=True)
    .replace(r"\,", ".", regex=True)
    .str.split()
)
dft.loc[dft["municipality_area_km2"].notna(), "municipality_area_km2"] = (
    dft
    .loc[dft["municipality_area_km2"].notna(), "municipality_area_km2"]
    .apply(lambda m: m[0])
    .apply(float)
)
dft
```
